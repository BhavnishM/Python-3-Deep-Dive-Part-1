```python
class Fib:
    def __init__(self, n):
        self.n = n
    def __len__(self):
        return self.n
    def __getitem__(self, value):
        if isinstance(value, int):
            if value<0:
                value = self.n + value
            elif value>=self.n:
                raise IndexError
            return self._fib(value)
        else:
            start, stop, step = value.indices(self.n)
            rng = range(start, stop, step)
            return [self._fib(i) for i in rng]
    @staticmethod
    @lru_cache(2*10)
    def _fib(n):
        if n<2:
            return 1
        else:
            return Fib._fib(n-1) + Fib._fib(n-2)
```

```python
class MyClass:
    def __init__(self, name):
        self.name = name
    def __repr__(self):
        return f'MyClass(name = {self.name})'
    def __add__(self, other):
        return MyClass(self.name + other.name)
    def __iadd__(self, other):
        if isinstance(other, MyClass):
            self.name += other.name
        else:
            self.name += other
        return self
    def __mul__(self, other):
        return MyClass(self.name * other)
    def __imul__(self, other):
        self.name *= other
        return self
    def __rmul__(self, other):
        return self.__mul__(other)
    def __contains__(self, other):
        return other in self.name
```

```python
from functools import total_ordering
import math
import numbers
@total_ordering
class Polygon:
    def __init__(self, E, R):
        if E>= 3 and isinstance(E, int) and isinstance(R, numbers.Number) and type(R) is not complex:
            self.edge = E
            self.circumradius = R
            self._edge_length = None
            self._vertices = None
            self._interior_angle = None
            self._apothem = None
            self._area = None
            self._perimeter = None
        else:
            raise ValueError('Number of edges must be greater than or equal to 3 and must be integer also circumradius must be a valid number')          
    @property
    def edge_length(self):
        if self._edge_length is None:
            print('calculating area')
            self._edge_length =  math.sin(math.pi/self.edge) * 2 * self.circumradius
        return self._edge_length
    @property
    def vertices(self):
        if self._vertices is None:
            self._vertices = self.edge
        return self._vertices
    @property
    def interior_angle(self):
        if self._interior_angle is None:
            self._interior_angle = (self.edge - 2) * 180/self.edge
        return self._interior_angle
    @property
    def apothem(self):
        if self._apothem is None:
            self._apothem = math.cos(math.pi/self.edge) * self.circumradius
        return self._apothem
    @property
    def area(self):
        if self._area is None:
            self._area = self.edge_length * self.apothem * self.edge / 2
        return self._area
    @property
    def perimeter(self):
        if self._perimeter is None:
            self._perimeter = self.edge * self.edge_length
        return self._perimeter
    def __repr__(self):
        return f'Polygon(edge = {self.edge}, circumradius = {self.circumradius})'
    def __eq__(self, other):
        if isinstance(other, Polygon):
            return (self.edge == other.edge and math.isclose(self.circumradius , other.circumradius, abs_tol = 0.001, rel_tol = 0.001))
        else:
             return NotImplemented
    def __gt__(self, other):
        if isinstance(other, Polygon):
            return self.vertices > other.vertices
        else:
            return NotImplemented
```
